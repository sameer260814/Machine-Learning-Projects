```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import sys
print('Python: {}'.format(sys.version))
```

```python
#importing the iris dataset
from sklearn import datasets
iris=datasets.load_iris()
features=iris.data
target=iris.target

df=pd.DataFrame(features)
df.columns=iris.feature_names
df.head()
```

```python
#dataset description
df.describe()
```

```python
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt

#displaying scatter plot matrix

scatter_matrix(df)
plt.show()
```

```python
#elbow method to determine no of clusters
from sklearn.cluster import KMeans
X=[]
Y=[]

for i in range(1,31):
    #initialise and fit the k means cluster
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df)
    
    #append no of cluster to x dataset
    X.append(i)
    
    #append avg within cluster sum of square (variance) to y list
    variance=kmeans.inertia_ / df.shape[0]
    Y.append(variance)
```

```python
import matplotlib.pyplot as plt
#plotting x and y
plt.plot(X,Y,'bo-')
plt.xlim(1,30)
plt.xlabel('No of Clusters')
plt.ylabel('Variance')
plt.title('Elbow Method')
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn import preprocessing
```

# PERFORMING PCA

```python
#perform PCA
pca=PCA(n_components=2)
pc=pca.fit_transform(df)

#printing new dimension
print(pc.shape)
print(pc[:10])
```

```python
#re-fit kmeans model to the principal components
kmeans=KMeans(n_clusters=3)
kmeans.fit(pc)
```

# Visualisation

```python
#setting mesh size
h=.02

#generate mesh grid
x_min, x_max=pc[:, 0].min() - 1, pc[:,0].max() + 1
y_min, y_max=pc[:, 1].min() - 1, pc[:,0].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

#label each point in mesh
Z = kmeans.predict(np.c_[xx.ravel(), yy.ravel()])

#generate color plot from resuts
Z = Z.reshape(xx.shape)
plt.figure(figsize=(10,10))
plt.clf() #clearing any past data that u have in plt
plt.imshow(Z, interpolation="nearest", extent=(xx.min(), xx.max(), yy.min(), yy.max()),
          aspect='auto', cmap=plt.cm.tab10, origin='lower')

#plot the principal component in color plot
for i, point in enumerate(pc):
    if target[i] == 0:
        plt.plot(point[0], point[1], 'g.', markersize=10)
    if target[i] == 1:
        plt.plot(point[0], point[1], 'b.', markersize=10)
    if target[i] == 2:
        plt.plot(point[0], point[1], 'r.', markersize=10)

#plot the cluster centroid
centroids=kmeans.cluster_centers_
plt.scatter(centroids[:,0], centroids[:,1], marker='x', s=250, linewidth=4, color='w', zorder=10)

#set plot title and axis limit
plt.title('Kmeans on iris dataset using PCA')
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.xticks(())
plt.yticks(())

#display
plt.show()
```

```python
#Metrics Perfromance
from sklearn import metrics

#kmeans on non-reduced data
kmeans1 = KMeans(n_clusters=3)
kmeans1.fit(features)

#kmeans on PCA-reduced data
kmeans2 = KMeans(n_clusters=3)
kmeans2.fit(pc)

#metric score for non_reduced data
print("Non_reduced data: ")
print('Homogeneity: {}'.format(metrics.homogeneity_score(target, kmeans1.labels_)))
print('Completeness: {}'.format(metrics.completeness_score(target, kmeans1.labels_)))
print('V-measure: {}'.format(metrics.v_measure_score(target, kmeans1.labels_)))
print()
#metric score for reduced data
print("PCA-reduced data: ")
print('Homogeneity: {}'.format(metrics.homogeneity_score(target, kmeans2.labels_)))
print('Completeness: {}'.format(metrics.completeness_score(target, kmeans2.labels_)))
print('V-measure: {}'.format(metrics.v_measure_score(target, kmeans2.labels_)))
```
